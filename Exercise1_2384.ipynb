```python
import matplotlib.pyplot as plt
import numpy as np

# Coefficients for the quadratic equation
a, b, c = 1, -3, 2

# Function to calculate the quadratic solution
def quadratic_solution(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant >= 0:
        root1 = (-b + discriminant**0.5) / (2*a)
        root2 = (-b - discriminant**0.5) / (2*a)
        return root1, root2
    else:
        return None  # Complex roots

# Calculate the roots
roots = quadratic_solution(a, b, c)

# Generate x and y values for the quadratic curve
x_values = np.linspace(-1, 4, 500)  # Adjust range to include roots
y_values = a * x_values**2 + b * x_values + c

# Plot the quadratic curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'{a}xÂ² + {b}x + {c}', color='blue')

# Highlight the roots, if they exist
if roots:
    root1, root2 = roots
    plt.scatter([root1, root2], [0, 0], color='red', label=f'Roots: x={root1:.2f}, x={root2:.2f}')

# Add x-axis and y-axis
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')

# Add labels, title, and legend
plt.title("Graph of the Quadratic Equation and Roots")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given coefficients for the quadratic equation
a, b, c = 1, -3, 2

# Define the quadratic equation as a function
def quadratic(x):
    return a * x**2 + b * x + c

# Generate x values for the graph (a range of values)
x_vals = np.linspace(-10, 10, 400)
y_vals = quadratic(x_vals)

# Plot the quadratic function
plt.plot(x_vals, y_vals, label="y = x^2 - 3x + 2")
plt.scatter(5, quadratic(5), color="red", zorder=5)  # Highlighting the point for x = 5
plt.text(5, quadratic(5), f"  (5, {quadratic(5)})", color="red")
plt.title("Graph of the Quadratic Equation")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
